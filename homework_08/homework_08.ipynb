## Задание 1.1

$A =
\begin{pmatrix}
2 & -4 \\
3 & 5 \\
-1 & 0
\end{pmatrix}
,
B =
\begin{pmatrix}
1 & 2 & 7 \\
-3 & -4 & 0 \\
5 & 2 & 1
\end{pmatrix}
,
C =
\begin{pmatrix}
6 & -3 & 9 \\
4 & -5 & 2 \\
8 & 1 & 5
\end{pmatrix}
.$

$D = A^T C - 2A^T B^T$

### Решение:

$A^T =
\begin{pmatrix}
2 & 3 & -1 \\
-4 & 5 & 0
\end{pmatrix}$,

$A^T * C =
\begin{pmatrix}
2 & 3 & -1 \\
-4 & 5 & 0
\end{pmatrix}
*
\begin{pmatrix}
6 & -3 & 9 \\
4 & -5 & 2 \\
8 & 1 & 5
\end{pmatrix}
=
\begin{pmatrix}
2*6+3*4-1*8 & ... & ... \\
... & ... & ...
\end{pmatrix}
=
\begin{pmatrix}
16 & -22 & 19 \\
-4 & -13 & -26
\end{pmatrix}$,

$2A^T =
\begin{pmatrix}
4 & 6 & -2 \\
-8 & 10 & 0
\end{pmatrix}
,
B^T =
\begin{pmatrix}
1 & -3 & 5 \\
2 & -4 & 2 \\
7 & 0 & 1
\end{pmatrix}$,

$2A^T * B^T =
\begin{pmatrix}
2 & -36 & 30 \\
12 & -16 & -20
\end{pmatrix}$

$D = A^T C - 2A^T B^T =
\begin{pmatrix}
14 & 14 & -11 \\
-16 & 3 & -6
\end{pmatrix}$

## Задание 1.2

$3 *
\begin{pmatrix}
x & 2 & 3 \\
-1 & y & 4
\end{pmatrix}
+ 2 *
\begin{pmatrix}
1 & 2 & -5 \\
2 & -6 & z
\end{pmatrix}
=
\begin{pmatrix}
8 & v & -1 \\
1 & 6 & 4
\end{pmatrix}$

### Решение

$\begin{pmatrix}
3x+2 & 10 & -1 \\
1 & 3y-12 & 12+2z
\end{pmatrix}
=
\begin{pmatrix}
8 & v & -1 \\
1 & 6 & 4
\end{pmatrix}$,

$\begin{matrix}
3x+2=8, & 3y-12=6, & 12+2z=4, & v=10 \\
x=2 & y=6 & z=-4
\end{matrix}$
