# Задание 1.5



```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
A = np.random.rand(n, n)
U, s, Vt = np.linalg.svd(A)
S = np.diag(s)

errors = list([np.linalg.norm(A - (U[:, :r] @ S[:r, :r] @ Vt[:r, :])) for r in range(2,n+1)])

plt.plot(range(2, n + 1), errors, marker='o')
plt.xlabel("r (кол-во сингулярных компонент)")
plt.ylabel("Ошибка восстановления")
plt.grid(True)
plt.show()
```

# Задание 2.4

### Решение

$f(x) = 1/3 (x^T x)^{3/2}$

По правилу цепочки функций $\frac{\partial f}{\partial x_k} = 1/3 * 3/2 (x^T x)^{1/2} * 2x_k = (x^T x)^{1/2} x_k$

Отсюда $\nabla f(x) = (x^T x)^{1/2} x = \|x\|_2 x$

Т.к. $df = \nabla f(x)^T dx$, то $df = \|x\|_2 x^T dx$, ($\|x\|_2$ - скаляр)


# Задание 2.5

### Решение

$f(x) = (x^T x)^{1/2}$

По правилу цепочки функций $\frac{\partial f}{\partial x_k} = 1/2 (x^T x)^{-1/2} * 2x_k = x_k / (x^T x)^{1/2}$

Отсюда $\nabla f(x) = x / (x^T x)^{1/2} = x / \|x\|_2$

Т.к. $df = \nabla f(x)^T dx$, то $df = (x^T / \|x\|_2) dx$, ($1 / \|x\|_2$ - скаляр)

# Задание 2.6

### Решение

$f(x) = ((Ax)^T Ax)^{1/2}$

$(Ax)^T Ax = x^T A^T Ax$ - квадратичная форма

Градиент квадратичной формы $x^T Bx = 2Bx$

По правилу цепочки функций $\nabla f(x) = 1/2 (A^T x^T Ax)^{-1/2} * 2A^T Ax = A^T Ax / \|Ax\|_2$

Т.к. $df = \nabla f(x)^T dx$, то $df = (A(Ax)^T / \|Ax\|_2) dx$, ($\|Ax\|_2$ - скаляр)
