## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
days = int(input('Введите количество дней до отпуска: '))
holydays = (days // 7) * 2 + max(((days % 7)-5), 0)
print(f"Количество выходных дней до отпуска: {holydays}")
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
length = int(input('Введите длину плитки: '))
width = int(input('Введите ширину плитки: '))
volume = int(input('Введите размер плитки: '))
print(f"Можно отломить такой кусок от этой плитки шоколада: {((volume % length) == 0) or ((volume % width) == 0)}")
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
num = int(input('Введите число до 3999 включительно:   '))

# Собираем Тысячи
num_1000 = ((num % 10000) // 1000) * 'M'

# Собираем Сотни
num_100     = (num % 1000) // 100
num_100_13  = (num_100 < 4) * num_100 * 'C'
num_100_4   = (num_100 == 4) * 'CD'
num_100_5   = (num_100 == 5) * 'D'
num_100_68  = ((num_100 > 5) and (num_100 < 9)) * ('D' + (num_100 - 5) * 'C')
num_100_9   = (num_100 == 9) * 'CM'
num_100     = num_100_13 + num_100_4 + num_100_5 + num_100_68 + num_100_9

# Собираем Десятки
num_10      = (num % 100) // 10
num_10_13   = (num_10 < 4) * num_10 * 'X'
num_10_4    = (num_10 == 4) * 'XL'
num_10_5    = (num_10 == 5) * 'L'
num_10_68   = ((num_10 > 5) and (num_10 < 9)) * ('L' + (num_10 - 5) * 'X')
num_10_9    = (num_10 == 9) * 'XC'
num_10      = num_10_13 + num_10_4 + num_10_5 + num_10_68 + num_10_9

# Собираем Единицы
num_1      = num_1 = num % 10
num_1_13   = (num_1 < 4) * num_1 * 'I'
num_1_4    = (num_1 == 4) * 'IV'
num_1_5    = (num_1 == 5) * 'V'
num_1_68   = ((num_1 > 5) and (num_1 < 9)) * ('V' + (num_1 - 5) * 'I')
num_1_9    = (num_1 == 9) * 'IX'
num_1      = num_1_13 + num_1_4 + num_1_5 + num_1_68 + num_1_9

print("Число в римском представлении:        " + num_1000 + num_100 + num_10 + num_1)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
num = input('Введите значение: ')

# Положительное целое число - 1) все символы являются цифрами и 2) не начинается на "0"
# Отрицательное целое число - 1) есть один минус и 2) он в начале и 3) остальные символы цифры и 4) не начинается на "-0"
# Положительное вещественное число - 1) есть одна точка и 2) она не в конце и 3) не в начале и 4) остальные символы цифры и 5) либо начинается на "0.", либо не начинается на "0"
# Отрицательное вещественное число - 1) есть один минус и 2) он в начале и 3) есть одна точка и 4) она не в началее и 5) не в конце и 6) остальные символы цифры и 7) либо начинается на "-0.", либо не начинается на "-0"

cond_int_positive   = num.isdigit() and not(num.startswith('0'))
cond_int_negative   = num.count('-') == 1 and num.startswith('-') and num.replace('-', '').isdigit() and not(num.startswith('-0'))
cond_float_positive = num.count('.') == 1 and not(num.endswith('.')) and not(num.startswith('.')) and num.replace('.', '').isdigit() and (num.startswith('0.') or not(num.startswith('0')))
cond_float_negative = num.count('-') == 1 and num.startswith('-') and num.count('.') == 1  and not(num.startswith('-.')) and not(num.endswith('.')) and num.replace('-', '').replace('.', '').isdigit() and (num.startswith('-0.') or not(num.startswith('-0')))

if cond_int_positive:
    print('Это положительное целое число.')
elif cond_int_negative:
    print('Это отрицательное целое число.')
elif cond_float_positive:
    print('Это положительное вещественно число.')
elif cond_float_negative:
    print('Это отрицательное вещественно число.')
else:
    print('Это не число.')
```
